# Kql - render scatterchart

```python
%%kql 
let randn = () {rand() + rand() + rand() + rand() + rand() + rand() + rand() + rand() + rand() + rand() + rand() + rand() + rand() + rand()};
range c from 1 to 1000 step 1 
| limit 1000
| extend e1 = randn()
| extend e2 = randn()
| extend e3 = randn()
| extend e4 = randn()
| extend e5 = randn()
| project e1,e2,e3,e4,e5
| render scatterchart title="Michael's chart";
```
